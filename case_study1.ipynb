#Bank account management system


```python
balance = 0

def deposit(amount):
    global balance
    if amount > 0:
        balance += amount
        print("Amount deposited:", amount)
    else:
        print("Invalid amount")

def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount
        print("Amount withdrawn:", amount)
    else:
        print("Insufficient balance")

def check_balance():
    print("Current balance:", balance)

# Menu
while True:
    print("\n1.Deposit 2.Withdraw 3.Check Balance 4.Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        amt = int(input("Enter amount: "))
        deposit(amt)

    elif choice == 2:
        amt = int(input("Enter amount: "))
        withdraw(amt)

    elif choice == 3:
        check_balance()

    elif choice == 4:
        break

    else:
        print("Invalid choice")
```

#Weather Data Analyzer using Python

```python
def weather_condition(temp, humidity):
    if temp > 30 and humidity < 50:
        return "Sunny ☀️"
    elif humidity > 70:
        return "Rainy 🌧️"
    else:
        return "Cloudy ☁️"

def average(values):
    return sum(values) / len(values)

# Input data
temps = []
humidity = []

n = int(input("Enter number of days: "))

for i in range(n):
    t = float(input(f"Day {i+1} Temperature: "))
    h = float(input(f"Day {i+1} Humidity: "))
    temps.append(t)
    humidity.append(h)

# Average
avg_temp = average(temps)
avg_hum = average(humidity)

print("\nAverage Temperature:", avg_temp)
print("Average Humidity:", avg_hum)

# Condition
print("Weather Condition:", weather_condition(avg_temp, avg_hum))
```
